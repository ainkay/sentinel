# Risk Index Calculation and Scoring

Compute weighted composite risk scores combining multiple anomaly detection indicators.

## 1. Setup and Data Generation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../')

from src import statistical_analysis, risk_index, feature_engineering

np.random.seed(42)

# Generate synthetic network traffic
n = 500
normal = {
    'Duration': np.random.exponential(500, int(n*0.85)),
    'Packets': np.random.poisson(10, int(n*0.85)),
    'Bytes': np.random.normal(2000, 500, int(n*0.85))
}

anomaly = {
    'Duration': np.random.exponential(50, int(n*0.15)),
    'Packets': np.random.poisson(200, int(n*0.15)),
    'Bytes': np.random.normal(50000, 10000, int(n*0.15))
}

df_normal = pd.DataFrame(normal)
df_anomaly = pd.DataFrame(anomaly)
df = pd.concat([df_normal, df_anomaly], ignore_index=True).sample(frac=1, random_state=42).reset_index(drop=True)
true_labels = ['Normal'] * len(df_normal) + ['Anomalous'] * len(df_anomaly)
true_labels_shuffled = [true_labels[i] for i in range(len(df))]

print(f"Dataset: {len(df)} records")
print(f"Normal: {len(df_normal)}, Anomalous: {len(df_anomaly)}")
print(f"\nData preview:")
print(df.head())

## 2. Compute Individual Anomaly Scores

# Normalize data first
scaler = feature_engineering.StandardScaler()
data_scaled = scaler.fit_transform(df.values)

print("Scaled data shape:", data_scaled.shape)
print("Scaled data (first 5 rows):")
print(data_scaled[:5])

# Method 1: Z-score based deviation
zscore_scores = statistical_analysis.compute_deviation_scores(data_scaled)
print(f"\nZ-score deviation scores:")
print(f"  Shape: {zscore_scores.shape}")
print(f"  Min: {zscore_scores.min():.4f}")
print(f"  Max: {zscore_scores.max():.4f}")
print(f"  Mean: {zscore_scores.mean():.4f}")
print(f"  Example scores (first 10): {zscore_scores[:10]}")

## 3. Compute IQR Scores

# Method 2: IQR-based anomaly scores
iqr_scores = np.zeros(len(df))

for col_idx, col in enumerate(df.columns):
    lower, upper = statistical_analysis.iqr_bounds(df[col], multiplier=1.5)
    # Distance from bounds
    distance = np.maximum(0, df[col].values - upper)
    distance = np.maximum(distance, lower - df[col].values)
    iqr_scores += np.abs(distance)

# Normalize IQR scores
iqr_scores_norm = feature_engineering.StandardScaler().fit_transform(iqr_scores.reshape(-1, 1)).flatten()
iqr_scores_norm = np.maximum(iqr_scores_norm, 0)  # Clip to [0, inf)

print(f"IQR anomaly scores:")
print(f"  Min: {iqr_scores_norm.min():.4f}")
print(f"  Max: {iqr_scores_norm.max():.4f}")
print(f"  Mean: {iqr_scores_norm.mean():.4f}")
print(f"  Example scores (first 10): {iqr_scores_norm[:10]}")

## 4. Compute Weighted Composite Risk Index

# Combine indicators into composite risk score
indicators = {
    'zscore': zscore_scores,
    'iqr': iqr_scores_norm
}

# Equal weights
weights_equal = {'zscore': 0.5, 'iqr': 0.5}
risk_scores_equal = risk_index.compute_weighted_risk_index(indicators, weights_equal)

# Z-score biased (more sensitive to normal distribution deviations)
weights_zscore_bias = {'zscore': 0.7, 'iqr': 0.3}
risk_scores_zscore = risk_index.compute_weighted_risk_index(indicators, weights_zscore_bias)

# IQR biased (more robust to outliers)
weights_iqr_bias = {'zscore': 0.3, 'iqr': 0.7}
risk_scores_iqr = risk_index.compute_weighted_risk_index(indicators, weights_iqr_bias)

print("Risk Scores (0-100 scale):")
print(f"\nEqual weights (0.5/0.5):")
print(f"  Min: {risk_scores_equal.min():.2f}")
print(f"  Max: {risk_scores_equal.max():.2f}")
print(f"  Mean: {risk_scores_equal.mean():.2f}")
print(f"  Example (first 10): {risk_scores_equal[:10]}")

print(f"\nZ-Score biased (0.7/0.3):")
print(f"  Example (first 10): {risk_scores_zscore[:10]}")

print(f"\nIQR biased (0.3/0.7):")
print(f"  Example (first 10): {risk_scores_iqr[:10]}")

## 5. Categorize Risk Levels

# Categorize risk scores
risk_categories = risk_index.categorize_risk(risk_scores_equal)

# Count by category
category_counts = pd.Series(risk_categories).value_counts()
print("Risk Category Distribution:")
print(category_counts)

# Show examples of each category
print("\nExamples by category:")
for cat in ['LOW', 'MEDIUM', 'HIGH', 'CRITICAL']:
    indices = np.where(risk_categories == cat)[0]
    if len(indices) > 0:
        sample_idx = indices[0]
        print(f"\n{cat}:")
        print(f"  Index: {sample_idx}")
        print(f"  Risk Score: {risk_scores_equal[sample_idx]:.2f}")
        print(f"  Data: Duration={df.iloc[sample_idx, 0]:.1f}, Packets={df.iloc[sample_idx, 1]:.1f}")

## 6. Create Risk DataFrame

# Create result dataframe with risk information
result_df = risk_index.create_risk_dataframe(df, risk_scores_equal)

print("Result DataFrame with Risk Scores:")
print(result_df.head(20))

# Summary statistics
print("\nShape:", result_df.shape)
print("Columns:", result_df.columns.tolist())
print("\nRisk Score Statistics:")
print(result_df['risk_score'].describe())

## 7. Anomaly Summary Stats

# Generate summary statistics
summary = risk_index.anomaly_summary_stats(result_df)

print("Anomaly Summary Statistics:")
print(f"  Total records: {summary['total_records']}")
print(f"  Anomalies detected: {summary['anomalies_detected']}")
print(f"  Anomaly rate: {summary['anomaly_percentage']:.2f}%")
print(f"  Average risk score: {summary['average_risk_score']:.2f}")
print(f"  Max risk score: {summary['max_risk_score']:.2f}")
print(f"  Min risk score: {summary['min_risk_score']:.2f}")
print(f"\nBreakdown by category:")
print(f"  CRITICAL: {summary['critical_count']}")
print(f"  HIGH: {summary['high_count']}")
print(f"  MEDIUM: {summary['medium_count']}")
print(f"  LOW: {summary['low_count']}")

## 8. Get Flagged Anomalies

# Get top anomalies
flagged = risk_index.flagged_anomalies(result_df, threshold=50)

print(f"Flagged Anomalies (risk > 50): {len(flagged)} records")
print("\nTop 10 highest risk anomalies:")
print(flagged[['Duration', 'Packets', 'Bytes', 'risk_score', 'risk_category', 'is_anomaly']].head(10))

# Separate by threshold
critical = result_df[result_df['risk_category'] == 'CRITICAL']
high = result_df[result_df['risk_category'] == 'HIGH']

print(f"\nCritical anomalies: {len(critical)}")
print(f"High anomalies: {len(high)}")
print(f"Total HIGH+CRITICAL: {len(critical) + len(high)}")

## 9. Visualization: Risk Distribution

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Risk Index Analysis', fontsize=14, fontweight='bold')

# Distribution of risk scores
axes[0, 0].hist(risk_scores_equal, bins=40, color='steelblue', edgecolor='black', alpha=0.7)
axes[0, 0].axvline(25, color='yellow', linestyle='--', linewidth=2, label='LOW/MEDIUM threshold')
axes[0, 0].axvline(50, color='orange', linestyle='--', linewidth=2, label='MEDIUM/HIGH threshold')
axes[0, 0].axvline(75, color='red', linestyle='--', linewidth=2, label='HIGH/CRITICAL threshold')
axes[0, 0].set_xlabel('Risk Score (0-100)', fontweight='bold')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].set_title('Risk Score Distribution')
axes[0, 0].legend()
axes[0, 0].grid(axis='y', alpha=0.3)

# Risk by category (bar chart)
categories = ['LOW', 'MEDIUM', 'HIGH', 'CRITICAL']
counts = [summary['low_count'], summary['medium_count'], summary['high_count'], summary['critical_count']]
colors = ['green', 'yellow', 'orange', 'red']
axes[0, 1].bar(categories, counts, color=colors, edgecolor='black', alpha=0.7)
axes[0, 1].set_ylabel('Number of Records', fontweight='bold')
axes[0, 1].set_title('Risk Category Distribution')
axes[0, 1].grid(axis='y', alpha=0.3)
for i, (cat, count) in enumerate(zip(categories, counts)):
    axes[0, 1].text(i, count + 5, str(count), ha='center', fontweight='bold')

# Risk vs feature (Duration)
axes[1, 0].scatter(df['Duration'], risk_scores_equal, alpha=0.5, s=30)
axes[1, 0].set_xlabel('Flow Duration', fontweight='bold')
axes[1, 0].set_ylabel('Risk Score', fontweight='bold')
axes[1, 0].set_title('Risk Score vs Flow Duration')
axes[1, 0].grid(alpha=0.3)

# Risk vs feature (Bytes)
axes[1, 1].scatter(df['Bytes'], risk_scores_equal, alpha=0.5, s=30, color='purple')
axes[1, 1].set_xlabel('Bytes Sent', fontweight='bold')
axes[1, 1].set_ylabel('Risk Score', fontweight='bold')
axes[1, 1].set_title('Risk Score vs Bytes Sent')
axes[1, 1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

## 10. Detection Accuracy (if true labels available)

# Compare with true labels
predicted_anomalies = result_df['is_anomaly'].astype(int).values
true_anomalies = [1 if label == 'Anomalous' else 0 for label in true_labels_shuffled]
true_anomalies = np.array(true_anomalies)

# Confusion matrix
TP = np.sum((predicted_anomalies == 1) & (true_anomalies == 1))
FP = np.sum((predicted_anomalies == 1) & (true_anomalies == 0))
FN = np.sum((predicted_anomalies == 0) & (true_anomalies == 1))
TN = np.sum((predicted_anomalies == 0) & (true_anomalies == 0))

# Metrics
accuracy = (TP + TN) / len(predicted_anomalies)
precision = TP / (TP + FP) if (TP + FP) > 0 else 0
recall = TP / (TP + FN) if (TP + FN) > 0 else 0
f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

print("Detection Performance:")
print(f"\nConfusion Matrix:")
print(f"  TP (Correct Anomalies): {TP}")
print(f"  FP (False Alarms): {FP}")
print(f"  FN (Missed Anomalies): {FN}")
print(f"  TN (Correct Normal): {TN}")

print(f"\nMetrics:")
print(f"  Accuracy: {accuracy:.2%}")
print(f"  Precision: {precision:.2%}")
print(f"  Recall: {recall:.2%}")
print(f"  F1-Score: {f1:.3f}")

## Summary

### Risk Index Approach

**Advantages:**
- Combines multiple indicators for robust detection
- Flexibility in weighting methods
- Scores 0-100 for easy interpretation
- Categories (LOW, MEDIUM, HIGH, CRITICAL) for action prioritization

**Key Components:**
1. **Z-Score Deviation**: Identifies points far from mean
2. **IQR Scores**: Robust to extreme outliers
3. **Weighted Combination**: Best of both worlds
4. **Risk Categories**: Actionable classifications

**SENTINEL Implementation:**
- Tuned weights: 50% Z-Score + 50% IQR
- Adjustable thresholds for different use cases
- Real-time scoring on streaming data
- Detailed auditability of decisions

This approach ensures both *sensitivity* to real attacks and *specificity* to minimize false alarms!
