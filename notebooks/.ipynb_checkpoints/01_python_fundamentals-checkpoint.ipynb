# Python Fundamentals for Data Science

Learn core Python concepts: loops, functions, conditionals, and data structures for anomaly detection.

## 1. Variables and Data Types

```python
# Numbers
risk_score = 75.5
anomaly_count = 15
is_anomaly = True

print(f"Risk Score: {risk_score}")
print(f"Anomaly Count: {anomaly_count}")
print(f"Is Anomaly: {is_anomaly}")
print(f"Type of risk_score: {type(risk_score)}")
```

## 2. Conditional Statements

```python
# Categorize risk based on score
risk_score = 65.0

if risk_score > 75:
    category = "HIGH"
elif risk_score > 50:
    category = "MEDIUM"
else:
    category = "LOW"

print(f"Risk Score: {risk_score} → Category: {category}")
```

## 3. Loops - For Loops

```python
# Process multiple risk scores
risk_scores = [20, 45, 78, 35, 88, 55, 92]
high_risk_count = 0

print("Risk Analysis:")
for score in risk_scores:
    if score > 75:
        high_risk_count += 1
        print(f"  Score {score}: HIGH RISK ⚠️")
    else:
        print(f"  Score {score}: Normal")

print(f"\nTotal High Risk Records: {high_risk_count}")
```

## 4. Loops - While Loops

```python
# Simulate anomaly detector threshold tuning
threshold = 50
detected = 0

while threshold <= 80 and detected < 10:
    # Simulate detection
    scores = [45, 75, 62, 88, 55, 92, 70, 80, 60, 90]
    detected = sum(1 for s in scores if s > threshold)
    print(f"Threshold: {threshold}, Detected Anomalies: {detected}")
    threshold += 5
```

## 5. Functions

```python
def categorize_risk(score):
    """Categorize risk score into levels."""
    if score >= 75:
        return "CRITICAL"
    elif score >= 50:
        return "HIGH"
    elif score >= 25:
        return "MEDIUM"
    else:
        return "LOW"

# Test the function
test_scores = [15, 45, 65, 85]
for score in test_scores:
    category = categorize_risk(score)
    print(f"Score: {score:3d} → {category}")
```

## 6. Function with Multiple Return Values

```python
def analyze_anomalies(scores):
    """Analyze array of anomaly scores."""
    total = len(scores)
    anomalies = sum(1 for s in scores if s > 75)
    average = sum(scores) / total
    
    return total, anomalies, average

# Get results
scores = [20, 45, 78, 35, 88, 55, 92, 65, 72, 15]
total, anomalies, average = analyze_anomalies(scores)

print(f"Total records: {total}")
print(f"Anomalies detected: {anomalies}")
print(f"Average score: {average:.2f}")
```

## 7. Lists and List Comprehensions

```python
# Traditional approach
scores = [20, 45, 78, 35, 88, 55, 92]
high_scores = []
for score in scores:
    if score > 75:
        high_scores.append(score)

print("Traditional:", high_scores)

# List comprehension (more Pythonic)
high_scores_comp = [s for s in scores if s > 75]
print("Comprehension:", high_scores_comp)

# Transform with comprehension
risk_categories = [categorize_risk(s) for s in scores]
print("Categories:", risk_categories)
```

## 8. Dictionaries

```python
# Store network flow information
flow = {
    'source_ip': '192.168.1.100',
    'destination_ip': '10.0.0.50',
    'port': 443,
    'bytes_sent': 2048,
    'packets': 15,
    'is_anomaly': False
}

print(f"Source: {flow['source_ip']}")
print(f"Bytes Sent: {flow['bytes_sent']}")
print(f"\nAll flow data:")
for key, value in flow.items():
    print(f"  {key}: {value}")
```

## 9. Default Parameter Values

```python
def compute_risk(zscore, iqr_score, zscore_weight=0.5, iqr_weight=0.5):
    """Compute weighted risk from multiple indicators."""
    risk = (zscore * zscore_weight + iqr_score * iqr_weight) / (zscore_weight + iqr_weight)
    return risk * 100  # Scale to 0-100

# Using defaults
risk1 = compute_risk(2.5, 1.8)
print(f"Risk (equal weights): {risk1:.1f}")

# Custom weights
risk2 = compute_risk(2.5, 1.8, zscore_weight=0.7, iqr_weight=0.3)
print(f"Risk (custom weights): {risk2:.1f}")
```

## Summary

- **Variables**: Store data with types (int, float, bool, str)
- **Conditionals**: if/elif/else for decision-making
- **Loops**: for and while for iteration
- **Functions**: Reusable code blocks with parameters and returns
- **Collections**: Lists, dicts for organizing data
- **Comprehensions**: Concise, Pythonic syntax

These fundamentals enable efficient anomaly detection algorithms!
